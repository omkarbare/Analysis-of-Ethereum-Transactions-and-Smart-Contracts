```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Open the text file
with open('transactions_total.txt', 'r') as f:
    # Read the file contents and split them into lines
    data = f.read()
```

```python
data = eval(data)
```

```python
df = pd.DataFrame(data, columns=["Date", "transactions_per_month"])

# Display the DataFrame
df.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('Month/Year')
plt.ylabel('Total Transactions each month')
plt.title('Bar plot of the number of transactions occurring every month')
x = df['Date']
y = df['transactions_per_month']
plt.bar(x, y, color='c')
plt.savefig('parta.png')
```
