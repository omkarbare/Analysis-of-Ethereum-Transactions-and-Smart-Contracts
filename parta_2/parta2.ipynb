```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Open the text file
with open('transactions_average.txt', 'r') as f:
    # Read the file contents and split them into lines
    data = f.read()
```

```python
data = eval(data)
data[0]
```

```python
dates = []
numbers = []
for i in data:
    my_list = i.split(',')
    date = my_list[0]
    dates.append(date)
    number = float(my_list[1])
    numbers.append(number)
```

```python
print(dates[0], numbers[0])
```

```python
df = pd.DataFrame({'dates': dates, 'average_transactions': numbers})
df.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('Month/Year')
plt.ylabel('Average Transactions each month')
plt.title('Bar plot of average value of transaction in each month')
x = df['dates']
y = df['average_transactions']
plt.bar(x,y,color='c')
plt.savefig('parta_2.png')
```
